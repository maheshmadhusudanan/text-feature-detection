```python
import cv2
import numpy as np
import imutils
import os
import math
import shutil
from matplotlib import pyplot as plt
```

```python
# Set figure width to 12 and height to 9
fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 19
fig_size[1] = 16
plt.rcParams["figure.figsize"] = fig_size
```

```python
image = cv2.imread("sample-1.jpg")
denoise = True
plt.imshow(image)
```

```python
edges = cv2.Canny(image,100,200)
```

```python
plt.subplot(121)
plt.imshow(image,cmap = 'gray')
```

```python
gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
edges = cv2.Canny(gray,50,150,apertureSize = 3)
minLineLength = 200
maxLineGap = 10
lines = cv2.HoughLinesP(edges,1,np.pi/180,100,minLineLength,maxLineGap)
for line in lines:
    for x1,y1,x2,y2 in line:
        cv2.line(image,(x1,y1),(x2,y2),(0,255,0),2)
cv2.imwrite('houghlines5.jpg',image)
```

```python
image = cv2.imread("acord-25-form-fillable-free.jpg",0)
thimg = cv2.adaptiveThreshold(image,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,11,2)
plt.imshow(thimg)
```
